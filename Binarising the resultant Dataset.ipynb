```python
def func(threshold_val):
    for i in df.index:
        count = count + 1
        for j in df.index:
      
            if(df.iloc[i,j] >=threshold_val):
                forumla = 1
                val.loc[i,j] = forumla
                
                        
            elif(df.iloc[i,j]<threshold_val):
                
                forumla = 0
                val.loc[i,j] = forumla                                       
```
