## Creating a 1000 X 1000 matrix just so we can store the distance values for Each Entry vs Each Entry ( 1000 Entries are present ! )

```python
def func():
    for i in df.index:
        
        for j in df.index:
        
            x14 = y11 = 10 - df.iloc[i,0]      #Adaptibility
            x11 = y12 = 10 - df.iloc[i,1]      #Ambition
            x12 = y13 = 10 - df.iloc[i,2]      #Desicveness
            x13 = y14 = 10 - df.iloc[i,3]      #Dominance
            x15 = y15 = 10 - df.iloc[i,4]      #Emotional
            x16 = y16 = 10 - df.iloc[i,5]      #faith
        
            x24 = y21 = df.iloc[j,0]      #Adaptibility
            x21 = y22 = df.iloc[j,1]      #Ambition
            x22 = y23 = df.iloc[j,2]      #Desicveness
            x23 = y24 = df.iloc[j,3]      #Dominance
            x25 = y25 = df.iloc[j,4]      #Emotional
            x26 = y26 = df.iloc[j,5]      #faith
        
            forumla = (y11-y21)*(y11-y21) + (y12-y22)*(y12-y22) + (y13-y23)*(y13-y23)+ (y14-y24)*(y14-y24) + (y15-y25)*(y15-y25) + (y16-y26)*(y16-y26); 
            val.loc[i,j] =  forumla**(1/2.0);
          
```

# Definining a function to calculate all possible values of the distances.

## We have defined the Distance Metric as Euclidean Distance in a 6 Dimensional Space

## The degree of closness is estimated by calculating the exact inverse of the entry present , and then finding the distance of the Entry being compared from the inverse of the First Entry.

To Give a better idea:

Entry 1: -->   10,5,6,7,2,3

Formula for calculating inverse of the entry :------> ( 10 - attribute_value)

therefore the inverse of Entry 1 :-----> (10-10)= 0, (10-5) = 5, (10-6) = 4, (10 - 7) = 3, (10 - 2) = 8, (10 - 3) = 7

i.e. Inverse(Entry 1) : ----->  0,5,4,3,8,7
     Orignal(Entry 1) : -----> 10,5,6,7,2,3
     
     Now we calculate the Euclidean Distance of Entry 2 from Inverse(Entry 1) on the 6 Dimensional Vector Space
     
          Euclidean Distance :------>     SquareRoot(Sum(Square(Attribute(Inverse(Entry1))-Attribute(Entry 2))))
     

```python
import time
start_time = time.time()
print(start_time)

func()

val.to_csv('Utility_val.csv')

end_time = time.time()
print("--- %s seconds ---" % ( end_time - start_time))
```
